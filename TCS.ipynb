```python
arr = input()
cnt_for_hash =0
cnt_for_star=0
for i in range(len(arr)):
    if arr[i] =="#":
        cnt_for_hash+=1
    elif arr[i] =="*":
        cnt_for_star+=1
if cnt_for_hash == cnt_for_star:
    print("0")
elif cnt_for_star>cnt_for_hash:
    print(cnt_for_star - cnt_for_hash)      
else:
    print(cnt_for_hash - cnt_for_star)              
   
```
