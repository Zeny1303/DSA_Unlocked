## Implementation of Queues using Arrays

```python
class Queue:
    def __init__(self):
        self.items=[]
    def __str__(self):
        return str(self.items)    
    def enqueue(self,x):
        self.items.append(x)
    def dequeue(self):
        if len(self.items)!=0:
            return self.items.pop(0)
        return "Queue Empty hai !!!"  
    def front(self):
        if len(self.items)!=0:
            return self.items[0]
        return "Queue Empty hai !!!"
    def rear(self):
        if len(self.items)!=0:
            return self.items[-1]
        return "Queue Empty hai !!!"
    def is_empty(self):
        return len(self.items)==0 #O(1)
    def size(self):
        return len(self.items)
    

    
queue  = Queue()
#insertion at end 
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(40)
queue.enqueue(60)

print(f"Queue contains {queue}")
# deletion at first 
print(f"Popped element is {queue.dequeue()}")
print(f"Popped element is {queue.dequeue()}")
print(f"front element of queue is {queue.front()}")
print(f"Rear(last) element of queue is {queue.rear()}")
print(f"Queue is empty = {queue.is_empty()}")
print(f"the size or length of queue is {queue.size()}")


```
