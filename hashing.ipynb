```python
n = [5,3,2,2,1,5,5,7,5,10]
m=[10,111,1,9,5,67,2]

for num in n:
    count = 0
    for val in m:
        if num == val:
            count += 1
    print(num, count)
```

```python
hash_list=[0]*11
for num in n:
    hash_list[num] += 1
for num in m:
    if num<1 or num>10:
        print(num,0)
    else:
        print(num,hash_list[num])
```

```python
hash_map = {}
for i in range(len(n)):
    hash_map[n[i]] = hash_map.get(n[i], 0) + 1
for num in m:
    if num<1 or num>10:
        print(num,0)
    else:     
        print(num,hash_map.get(num,0))
    
```

```python
s = "A man, a plan, a canal: Panama"
# remove non-alphanumeric characters and convert to lowercase
s = ''.join(c.lower() for c in s if c.isalnum())
print(s)  # for debugging purpose
# check if the string is equal to its reverse
is_palindrome = s == s[::-1]
print(is_palindrome)
```

```python
def rotate(nums,k):
    for i in range(0,k):
        temp =nums[len(nums)-1]
        for j in range(len(nums)-2,-1,-1):
            nums[i+1]=nums[i]
        nums[0]=temp    
    return nums
nums=[1,2,3,4,5,6,7]
k=3
rotate(nums,k)    
```

```python
nums=[3,1,-2,-5,2,-4]
pos_nums=[]
neg_nums=[]
for i in range(0,len(nums)):
    if nums[i]>=0:
        pos_nums.append(nums[i])
    else:
        neg_nums.append(nums[i])
print(pos_nums)
print(neg_nums)          
```
